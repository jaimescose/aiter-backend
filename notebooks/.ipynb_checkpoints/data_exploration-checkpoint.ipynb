```python
import pandas as pd
```

```python
df = pd.read_csv('data/apple_dropped.csv')
```

```python
data_types = [
    'ActiveEnergyBurned',
    'AppleExerciseTime',
    'AppleStandTime',
    'BasalEnergyBurned',
    'BodyFatPercentage',
    'BodyMass',
    'HeartRate',
    'LeanBodyMass',
    'RespiratoryRate',
    'RestingHeartRate',
    'SleepAnalysis',
    'StepCount',
    'WalkingHeartRateAverage',
    'VO2Max',
    'DistanceWalkingRunning'
]
```

```python
df_dict = {t: df[df['type'] == t].copy() for t in data_types}
```

```python
for t, df_temp in df_dict.items():
    df_temp.to_csv(f'data/metrics/{t}.csv', index=False)
```

```python
df_extra = df[df['type'].isna()].copy()
df_extra.head()
```

```python
df_extra.to_csv('data/metrics/extra_info.csv', index=False)
```

```python
default_columns = ['type', 'value', 'unit', 'startDate', 'endDate']
```

```python
# walking-distance, stand-time,... to be continued
```

```python
for t, df_temp in df_dict.items():
    df_dict[t] = df_temp[default_columns]
```

```python
conversions_dict = {
    'DistanceWalkingRunning': {'unit': 'mi', 'new_unit': 'meter(s)', 'f': lambda x: x * 1609, 'dtype': float, 'indicator': 'sum'},
    'AppleStandTime': {'unit': 'min', 'new_unit': 'min', 'dtype': int, 'f': lambda x: x / 60, 'indicator': 'sum'},
    'HeartRate': {'unit': 'count/min', 'new_unit': 'count/min', 'dtype': float, 'indicator': 'mean'},
    'VO2Max' : {'unit': 'mL/min·kg', 'new_unit': 'mL/min·kg', 'dtype': float, 'indicator': 'mean'}
}
```

```python
metric = 'VO2Max'
unit = conversions_dict[metric]['unit']
new_unit = conversions_dict[metric]['new_unit']
dtype = conversions_dict[metric]['dtype']
indicator = conversions_dict[metric]['indicator']
f = conversions_dict[metric].get('f')
```

```python
df = df_dict[metric].copy()
```

```python
# Convert 'datetime' column to datetime dtype
# .loc[:, 'startDate']
df.loc[:, 'start_date'] = pd.to_datetime(df['startDate']).dt.date
df.loc[:, 'end_date'] = pd.to_datetime(df['startDate']).dt.date
df.loc[:, 'value'] = df['value'].astype(dtype)
```

```python
from datetime import datetime, timedelta

# Get the current date
current_date = datetime.now().date()  # Get current date without time component

# Calculate date 30 days back
date_30_days_back = current_date - timedelta(days=30)

# Calculate date 7 days back
date_7_days_back = current_date - timedelta(days=7)
```

```python
df = df[df['start_date'] >= date_30_days_back].copy()
window = 'month'
# df = df[df['start_date'] >= date_7_days_back].copy()
# window = 'week'
df
```

```python
# df_agg = df.groupby('start_date')['value'].sum().reset_index()
if indicator == 'mean':
    df_agg = df.groupby('start_date')['value'].mean().reset_index()
else:
    df_agg = df.groupby('start_date')['value'].sum().reset_index()
df_agg
```

```python
if f:
    df_agg[new_unit] = df_agg['value'].apply(f)
else:
    df_agg[new_unit] = df_agg['value']
```

```python
df_agg
```

```python
df_agg.to_csv(f'data/metrics/agg/{metric}_{window}.csv', index=False)
```
